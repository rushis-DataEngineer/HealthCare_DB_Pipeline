```python
silver_path = f"{RAW_PATH}/silver/{silver_catalog}"
print(silver_path)
```

```python

new_df = spark.read.format('parquet').options(header=True, inferSchema=True).load(silver_path)
```

```python
display(new_df.limit(50))
```

```python
from pyspark.sql import functions as F
```

```python
# Display the count of patients for each medical condition as a table for visualization
display(new_df.groupBy('Condition').agg(F.count('Patient Name').alias('Patient Count')).select('Condition', 'Patient Count'))
```

```python
# Extract the year from the 'Visit Date' column and add it as a new column 'Year'
new_df = new_df.withColumn('Year', F.year(F.col("Visit Date")))
```

```python
# Extract the year from the 'Visit Date' column and add it as a new column 'Year'
new_df = new_df.withColumn(
    'Year',
    F.year(F.col("Visit Date"))
)
```

```python
final_df = new_df.select("Patient Name", "Age","Gender", "Condition", "Medication", "Blood Pressure", "Cholesterol", "Visit Date", "Year")
```
